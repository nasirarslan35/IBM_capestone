### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
# SQL command to select unique launch sites
sql_command = """
SELECT DISTINCT "Launch_Site" FROM SPACEXTBL;
"""

# Execute the SQL command
cur.execute(sql_command)

# Fetch all results
launch_sites = cur.fetchall()

# Print each launch site
for site in launch_sites:
    print(site[0])

# Close the connection
#con.close()
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'KSC' 


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL command to select records where launch sites begin with 'KSC'
sql_command = """
SELECT * FROM SPACEXTBL
WHERE "Launch_Site" LIKE 'KSC%' 
LIMIT 5;
"""

# Execute the SQL command
cur.execute(sql_command)

# Fetch all results
records = cur.fetchall()

# Print each record
for record in records:
    print(record)

# Close the connection
con.close()
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL command to sum payload mass for launches by NASA (CRS)
sql_command = """
SELECT SUM("PAYLOAD_MASS__KG_") AS Total_Payload_Mass
FROM SPACEXTBL
WHERE "Customer" = 'NASA (CRS)';
"""

# Execute the SQL command
cur.execute(sql_command)

# Fetch the result
total_payload_mass = cur.fetchone()

# Check if the result is not None
if total_payload_mass[0] is not None:
    print("Total payload mass carried by boosters launched by NASA (CRS):", total_payload_mass[0], "kg")
else:
    print("No data available for NASA (CRS) launches or payload mass is not recorded.")

# Close the connection
con.close()
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL command to calculate the average payload mass for the booster version 'F9 v1.1'
sql_command = """
SELECT AVG("PAYLOAD_MASS__KG_") AS Average_Payload_Mass
FROM SPACEXTBL
WHERE "Booster_Version" = 'F9 v1.1';
"""

# Execute the SQL command
cur.execute(sql_command)

# Fetch the result
average_payload_mass = cur.fetchone()

# Check if the result is not None
if average_payload_mass[0] is not None:
    print("Average payload mass carried by booster version F9 v1.1:", round(average_payload_mass[0], 2), "kg")
else:
    print("No data available for booster version F9 v1.1 or payload mass is not recorded.")

# Close the connection
con.close()
```

### Task 5

##### List the date where the succesful landing outcome in drone ship was acheived.


_Hint:Use min function_ 


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL command to find the earliest date of a successful drone ship landing
sql_command = """
SELECT MIN("Date") AS Earliest_Successful_Landing
FROM SPACEXTBL
WHERE "Landing_Outcome" = 'Success (drone ship)';
"""

# Execute the SQL command
cur.execute(sql_command)

# Fetch the result
earliest_landing_date = cur.fetchone()

# Check if the result is not None
if earliest_landing_date[0] is not None:
    print("The earliest successful drone ship landing was on:", earliest_landing_date[0])
else:
    print("No successful drone ship landings recorded.")

# Close the connection
con.close()
```

### Task 6

##### List the names of the boosters which have success in ground pad  and have payload mass greater than 4000 but less than 6000


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL command to list booster names with successful ground pad landings and specific payload mass range
sql_command = """
SELECT "Booster_Version"
FROM SPACEXTBL
WHERE "Landing_Outcome" = 'Success (ground pad)'
AND "PAYLOAD_MASS__KG_" > 4000 AND "PAYLOAD_MASS__KG_" < 6000;
"""

# Execute the SQL command
cur.execute(sql_command)

# Fetch all results
boosters = cur.fetchall()

# Check if results were found
if boosters:
    print("Boosters with successful ground pad landings and payload mass between 4000 and 6000 kg:")
    for booster in boosters:
        print(booster[0])
else:
    print("No boosters found meeting the criteria.")

# Close the connection
con.close()
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL command to count successful and failed mission outcomes
sql_command = """
SELECT "Mission_Outcome", COUNT(*) AS Outcome_Count
FROM SPACEXTBL
WHERE "Mission_Outcome" LIKE '%Success%' OR "Mission_Outcome" LIKE '%Failure%'
GROUP BY "Mission_Outcome";
"""

# Execute the SQL command
cur.execute(sql_command)

# Fetch all results
outcomes = cur.fetchall()

# Print the outcomes
if outcomes:
    for outcome in outcomes:
        print(f"Outcome: {outcome[0]}, Count: {outcome[1]}")
else:
    print("No successful or failed mission outcomes found.")

# Close the connection
con.close()
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL command to list booster versions that have carried the maximum payload mass
sql_command = """
SELECT "Booster_Version"
FROM SPACEXTBL
WHERE "PAYLOAD_MASS__KG_" = (
    SELECT MAX("PAYLOAD_MASS__KG_")
    FROM SPACEXTBL
);
"""

# Execute the SQL command
cur.execute(sql_command)

# Fetch all results
booster_versions = cur.fetchall()

# Print the booster versions
if booster_versions:
    print("Booster versions that have carried the maximum payload mass:")
    for booster in booster_versions:
        print(booster[0])
else:
    print("No data found for maximum payload mass.")

# Close the connection
con.close()

```

### Task 9


##### List the records which will display the month names, succesful landing_outcomes in ground pad ,booster versions, launch_site for the months in year 2017
**Note: SQLLite does not support monthnames. So you need to use substr(Date,6,2) for month, substr(Date,9,2) for date, substr(Date,0,5),='2017' for year.**


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL command to list desired records for the year 2017 and successful landings on a ground pad
sql_command = """
SELECT 
    substr(Date, 6, 2) AS Month,
    "Booster_Version",
    "Launch_Site",
    "Landing_Outcome"
FROM SPACEXTBL
WHERE 
    substr(Date, 1, 4) = '2017' AND
    "Landing_Outcome" = 'Success (ground pad)';
"""

# Execute the SQL command
cur.execute(sql_command)

# Fetch all results
records = cur.fetchall()

# Check if results were found
if records:
    print("Month, Booster Version, Launch Site, Landing Outcome")
    for record in records:
        # Mapping the numeric month to month name, if needed
        month_number = int(record[0])
        month_name = ["January", "February", "March", "April", "May", "June",
                      "July", "August", "September", "October", "November", "December"][month_number - 1]
        
        print(f"{month_name}, {record[1]}, {record[2]}, {record[3]}")
else:
    print("No records found for 2017 with successful landings on a ground pad.")

# Close the connection
con.close()
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("my_data1.db")
cur = con.cursor()

# SQL command to count and rank landing outcomes between specified dates
sql_command = """
SELECT "Landing_Outcome", COUNT(*) AS Outcome_Count
FROM SPACEXTBL
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY "Landing_Outcome"
ORDER BY Outcome_Count DESC;
"""

# Execute the SQL command
cur.execute(sql_command)

# Fetch all results
outcomes = cur.fetchall()

# Check if results were found
if outcomes:
    print("Landing Outcome, Count")
    for outcome in outcomes:
        print(f"{outcome[0]}, {outcome[1]}")
else:
    print("No landing outcomes recorded between 2010-06-04 and 2017-03-20.")

# Close the connection
con.close()

```
