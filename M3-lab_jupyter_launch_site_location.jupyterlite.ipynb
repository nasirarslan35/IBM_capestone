```python
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Function to determine marker color
def choose_marker_color(launch_class):
    if launch_class == 1:
        return 'green'  # Success
    else:
        return 'red'  # Failure

# Apply the function to the DataFrame
spacex_df['marker_color'] = spacex_df['class'].apply(choose_marker_color)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
import folium
from folium.plugins import MarkerCluster

# Initialize map and marker cluster
nasa_coordinate = [29.559684888503615, -95.0830971930759]  # Example coordinates for NASA
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)
marker_cluster = MarkerCluster().add_to(site_map)

# Define marker color in spacex_df
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Add markers to the cluster
for index, row in spacex_df.iterrows():
    marker = folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=f"Launch Site: {row['Launch Site']}<br>Outcome: {'Success' if row['class'] == 1 else 'Failure'}",
        icon=folium.Icon(color='white', icon_color=row['marker_color'])
    )
    marker_cluster.add_child(marker)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
from folium.plugins import MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Define coordinates and map
launch_site_coordinates = [28.562302, -80.577356]
closest_city = [28.3922, -80.6077]   # Cape Canaveral
closest_railway = [28.5721, -80.5853]  # Example coordinates for railway
closest_highway = [28.5634, -80.5706]  # Example coordinates for highway

# Function to calculate distance using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # approximate radius of earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Calculate distances
distance_city = calculate_distance(launch_site_coordinates[0], launch_site_coordinates[1], closest_city[0], closest_city[1])
distance_railway = calculate_distance(launch_site_coordinates[0], launch_site_coordinates[1], closest_railway[0], closest_railway[1])
distance_highway = calculate_distance(launch_site_coordinates[0], launch_site_coordinates[1], closest_highway[0], closest_highway[1])

# Initialize map
site_map = folium.Map(location=launch_site_coordinates, zoom_start=12)

# Add markers
city_marker = folium.Marker(
    location=closest_city,
    icon=DivIcon(
        icon_size=(150, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_city),
    )
)
site_map.add_child(city_marker)

railway_marker = folium.Marker(
    location=closest_railway,
    icon=DivIcon(
        icon_size=(150, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_railway),
    )
)
site_map.add_child(railway_marker)

highway_marker = folium.Marker(
    location=closest_highway,
    icon=DivIcon(
        icon_size=(150, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_highway),
    )
)
site_map.add_child(highway_marker)

# Draw lines
city_line = folium.PolyLine(locations=[launch_site_coordinates, closest_city], weight=4, color='blue')
site_map.add_child(city_line)

railway_line = folium.PolyLine(locations=[launch_site_coordinates, closest_railway], weight=4, color='blue')
site_map.add_child(railway_line)

highway_line = folium.PolyLine(locations=[launch_site_coordinates, closest_highway], weight=4, color='blue')
site_map.add_child(highway_line)

# Add mouse position to the map for easier coordinate finding
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Show map
site_map

```
